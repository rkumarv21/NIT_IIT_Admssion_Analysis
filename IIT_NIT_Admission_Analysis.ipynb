## 1.0 Importing the intial Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df = pd.read_csv('iit_nit_data.csv')
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize = (8,6))
sns.heatmap(df.corr(), annot = True, data = df, linewidth = 2)
```

```python
for var in df.drop(['id','is_preparatory'], axis = 1 ):
    plt.figure(figsize = (10,6), dpi = 100)
    sns.countplot(var, hue = 'is_preparatory', data = df)
    plt.title(f'barplot of {var}')
```

```python
df_var = df[['year', 'institute_type', 'institute_short',
                 'round_no', 'quota','pool', 'program_duration', 
                 'degree_short', 'category',
                'is_preparatory']]
```

```python
for i in df_var:
    print(df_var[i].unique())
```

```python
for i in df_var:
    print(df_var[i].value_counts())
```

```python
for i in df_var:
    plt.figure(figsize = (10,6))
    sns.countplot(df_var[i], palette = 'hls', data = df_var)
    plt.xticks(rotation = 90)
```

```python
for i in df_var:
    plt.figure(figsize = (10,6))
    df_var[i].value_counts().plot(kind = 'pie')
    plt.title(f'pie chart of {i}.')
```

```python
plt.figure(figsize = (8,6), dpi = 100)
df_year_quota = df.groupby(['year','quota']).size().plot(kind = 'line', color = 'blue')
plt.xticks(rotation = 90)
```

```python
print(f'Average Opening rank {int(df.opening_rank.mean())}')
print(f'Minimum Opening rank {int(df.opening_rank.min())}')
print(f'Maximum Opening rank {int(df.opening_rank.max())}')
```

## 2.0 Analysis through Visualisation

```python
plt.figure(figsize = (10,8), dpi = 100)
df_opening_rank_year = df.groupby(['opening_rank', 'year']).size().plot(kind = 'line')
plt.xticks(rotation = 90)
```

```python
plt.figure(figsize = (10,8), dpi = 100)
df_closing_rank_year = df.groupby(['closing_rank', 'year']).size().plot(kind = 'line')
plt.xticks(rotation = 90)
```

```python
plt.figure(figsize = (10,8), dpi = 100)
df_opening_rank = df.groupby(['institute_short', 'is_preparatory']).is_preparatory.count().plot(kind = 'bar', color = 'green')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(15,8))
df_preparatory_round = df.groupby(['is_preparatory','round_no']).round_no.count().plot(kind = 'line',color = 'black')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(12,7))
df_preparatory_round = df.groupby(['is_preparatory','round_no']).round_no.count().plot(kind = 'bar',color = 'magenta')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(15,8))
df_degree_institute = df.groupby(['degree_short','institute_short']).institute_short.count().head(20).plot(kind = 'bar',color = 'blue')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(12,7))
df_program_institute = df.groupby(['program_duration','institute_short']).institute_short.count().head(10).plot(kind = 'bar',color = 'magenta')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(15,8))
df_category_preparatory = df.groupby(['category','is_preparatory']).is_preparatory.count().head(10).plot(kind = 'bar',color = 'green')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(15,8))
df_category_preparatory = df.groupby(['category','is_preparatory']).is_preparatory.count().head(10).plot(kind = 'line',color = 'black')
plt.xticks(rotation = 90)
plt.show()
```
